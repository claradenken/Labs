### For the numerical variables, check the multicollinearity between the input features.
#### (Multicollinearity does not include the target feature "total_claim_amount")

```python
nums_without_tf = nums.drop(labels="total_claim_amount",axis=1)
corr = nums_without_tf.corr()
corr
```

```python
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5), dpi=90)
    ax = sns.heatmap(round(corr,2), vmin=-1, vmax=1, cmap='coolwarm', mask=mask, annot=True, square=True)
    ax.set_title("Heatmap customer data", weight="bold", size=20)
```

- The highest correlation is between customer_lifetime_value and monthly_premium_auto (0.4).
- The other correlations are all very low (below an absolute value of 0.1).

#### Show a plot of the total number of responses.


```python
fig, ax = plt.subplots(figsize=(7,1))

resp = dict(cats.response.value_counts())
ax.barh(list(resp.keys()),list(resp.values()),color=["crimson","seagreen"])
ax.set_xticks(range(0,sum(resp.values()),1000))
ax.set_title("Number of responses by category", weight="bold")
ax.set_xlabel("number of responses", size=10, weight="bold")
ax.set_ylabel("answer", size=11, weight="bold")
plt.show()
```

#### Show a plot of the response by the sales channel.

```python
# this part is not necessary for plotting in this case (with countplot), only for understanding
scr = cats.groupby(["sales_channel","response"])["response"].count().rename("count").reset_index()
scr
```

```python
palette_alive = {"yes": "seagreen", "no": "crimson"}
fig, ax = plt.subplots(figsize=(8,5))
ax = sns.countplot(x = cats.sales_channel,  hue = cats.response, palette=palette_alive)
ax.set_xlabel("Sales channel")
ax.set_ylabel("Count")
plt.show()
```

#### Show a plot of the response by the total claim amount.

```python
df['rounded_claim_amount'] = df['total_claim_amount'].apply(round, ndigits = -2)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax = sns.countplot(x = df.rounded_claim_amount,  hue = df.response, palette=palette_alive)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_title("Response by total claim amount", weight="bold")
ax.legend(loc="upper center")
ax.set_xlabel("Total claim amount")
ax.set_ylabel("Count")
plt.show()
```

#### Show a plot of the response by income.

```python
df['rounded_income'] = df['income'].apply(round, ndigits = -4)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.countplot(x = df.rounded_income,  hue = df.response, palette=palette_alive)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_title("Response by income", weight="bold")
ax.legend(loc="upper center")
ax.set_xlabel("Income, rounded to the next ten thousands")
ax.set_ylabel("Count")
plt.show()
```
