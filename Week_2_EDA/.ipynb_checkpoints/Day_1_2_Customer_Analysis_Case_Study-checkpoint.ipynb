#### Replacing null values – Replace missing values with means of the column (for numerical columns). Pay attention that the Income feature for instance has 0s which is equivalent to null values. (We assume here that there is no such income with 0 as it refers to missing values) Hint: numpy.nan is considered of float64 data type.

```python
num_lst = ["Customer Lifetime Value","Income","Monthly Premium Auto","Number of Open Complaints","Total Claim Amount"]
df[num_lst].isna().sum()
df[num_lst].info()
```

```python
df["Income"] = [x if x!=0 else None for x in df["Income"]]
```

```python
def mean_for_na(col):
    mean_col = np.mean(df[col])
    df[col] = df[col].fillna(mean_col)
```

```python
for col in num_lst:
    mean_for_na(col)
```

#### Bucketing the data - Write a function to replace column "State" to different zones. California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central

```python
df.State.value_counts()
```

```python
def state_to_zone(state):
    if state=="California": return "West"
    elif state=="Oregon": return "North West"
    elif state=="Washington": return "East"
    elif state in ["Arizone","Nevada"]: return "Central"
```

```python
df.State = df["State"].apply(state_to_zone)
```

#### (Optional) Removing outliers using 1.5*IQR technique for all numerical columns.

```python
def outlier_treatment(col):
    sorted(col)
    Q1,Q3 = np.percentile(col, [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    lower_range
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
```

```python
col = df.Income
sorted(col)
Q1,Q3 = np.percentile(col, [25,75])
IQR = Q3 - Q1
lower_range = Q1 - (1.5 * IQR)
upper_range = Q3 + (1.5 * IQR)
lower_range
print(lower_range,IQR,upper_range)
```

```python
for col in num_lst:
    lowerbound,upperbound = outlier_treatment(df[col])
    outliers = df[col][(df[col].lt(lowerbound) | df[col].gt(upperbound))]      # lt: lower than, gt: greater than
    print(col,":",len(outliers),"outliers")
    df[col].drop(outliers.index , inplace=True)
```
